```python
import pandas as pd
```

```python
def float_or_false(value):
    try:
        if float(value):
            return True
    except ValueError:
        return False
```

```python
def mod_false(df):
    for i in range(len(df)):
        if df.loc[i,'modifications'] != '':
            df.loc[i,'is_mod'] = 'True'
        else:
            df.loc[i,'is_mod'] = 'False'
```

```python
df = pd.read_csv('serum_original.csv')
df=df.rename(columns={'Sequence':'sequence', 'Half-life (hours)': 'half_life_seconds', 'Modifications':'modifications'})
```

```python
df['modifications'] = df['modifications'].replace('N.A.', '')
mod_false(df)
```

```python
df['hl_is_float'] = df['half_life_seconds'].apply(float_or_false)
df_float=df[df['hl_is_float']==True]
df_object=df[df['hl_is_float']==False]
df_float.reset_index(drop=True, inplace=True)
df_object.reset_index(drop=True, inplace=True)
```

```python
for i in range (len(df_float)):
    df_float.loc[i,'half_life_seconds'] = float(df_float.loc[i,'half_life_seconds'])*3600
```

```python
df_object.reset_index(drop=True, inplace=True)
df_object
```

```python
df_float=df_float.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
df_object=df_object.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
```
