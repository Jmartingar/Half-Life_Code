```python
import pandas as pd
```

```python
def float_or_false(value):
    try:
        if float(value):
            return True
    except ValueError:
        return False
```

```python
def mod_false(df):
    for i in range(len(df)):
        if df.loc[i,'modifications'] != '':
            df.loc[i,'is_mod'] = 'True'
        else:
            df.loc[i,'is_mod'] = 'False'
```

```python
df = pd.read_excel('../raw/Estimating_Serum.xlsx', header=1)
df=df.rename(columns={'Sequence':'sequence', 'Half-life (hours)': 'half_life_seconds', 'Modifications':'modifications'})
```

```python
df_filtered = df[['sequence', 'half_life_seconds', 'modifications']]
df_filtered['modifications'] = df_filtered['modifications'].replace('N.A.', '')
mod_false(df_filtered)
```

```python
df_filtered['hl_is_float'] = df_filtered['half_life_seconds'].apply(float_or_false)
df_float=df_filtered[df_filtered['hl_is_float']==True]
df_object=df_filtered[df_filtered['hl_is_float']==False]
```

```python
df_filtered
```

```python
df_float.loc[:, 'half_life_seconds'] = df_float['half_life_seconds'] * 3600
df_float.reset_index(drop=True, inplace=True)
df_float
```

```python
df_object.reset_index(drop=True, inplace=True)
df_object
```

```python
df_float = df_float.drop(['hl_is_float'], axis=1)
df_float.to_csv('../Serum/serum.csv', index=False)
df_object = df_object.drop(['hl_is_float'], axis=1)
df_object.to_csv('../Serum/serum_object.csv', index=False)
```
