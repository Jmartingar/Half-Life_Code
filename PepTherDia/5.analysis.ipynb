```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_orig = pd.read_csv("PepTherDia.csv")
df_mod = pd.read_csv("PepTherDia_mod.csv")
df_nomod = pd.read_csv("PepTherDia_nomod.csv")
#estos son a partir de los no modificados
df_canon = pd.read_csv("PepTherDia_nomod_canon.csv")
df_nocanon = pd.read_csv("PepTherDia_nomod_notcanon.csv")
```

```python
sns.histplot(df_mod['half_life_seconds'], bins=10, color='skyblue', kde=False)

plt.xlabel('Vida media (segundos)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Vida Media (PepTherDia)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='half_life_seconds', data=df_mod)
plt.title('Boxplot de Vida Media (PepTherDia)')
plt.xlabel('Vida media (segundos)')
plt.show()
```
