```python
import pandas as pd
import numpy as np
```

```python
def mod_col(df):
    for i in range(len(df)):
        df.loc[i,'modifications'] = f"{df.loc[i, 'N-terminus']}, {df.loc[i, 'C-terminus']}"
        df.loc[i,'is_mod'] = 'True'
```

```python
df = pd.read_csv('PepTherDia_original.csv')
```

```python
df.drop(columns=['ID'], inplace=True)
```

```python
df2 = df
mod_col(df2)
```

```python
df2.drop(['N-terminus', 'C-terminus'], axis=1, inplace=True)
df2.rename(columns={'Sequence': 'sequence', 'Terminal Half-Life': 'half_life_seconds'}, inplace=True)
df2['sequence'] = df2['sequence'].str.upper()
```

```python
#dimensiones de datos duplicados
df2[df2.duplicated(subset=['sequence', 'modifications'], keep=False)].shape
```

```python
#dimensiones de datos no duplicados
df2[~df2.duplicated(subset=['sequence', 'modifications'], keep=False)].shape
```

```python
#separacion de datos duplicados que tengan la misma secuencia y modificaciones (se examinan las secuencia, si modificaiones y secuencia es igual pero half_life_seconds es diferente se borran todos los duplicados)
df_data_duplicated=(df2[df2.duplicated(subset=['sequence', 'modifications'], keep=False)]).sort_values(by='sequence')
df_data_duplicated
```

```python
#separacion de datos duplicados que tengan la misma secuencia, modificaciones y half_life_seconds (de esto solo se guarda una copia) 
df_data_duplicated_2=(df2[df2.duplicated(subset=['sequence', 'modifications', 'half_life_seconds'], keep=False)]).sort_values(by='sequence')
df_data_duplicated_2
```

```python
#separacion de datos no duplicados
df_not_duplicated=(df2[~df2.duplicated(subset=['sequence', 'modifications'], keep=False)]).sort_values(by='sequence')
#se concatenan los datos no duplicados con los datos duplicados pero que se mantiene una copia
#PREGUNTAR POR ESTOS DUPLICADOS EN DUPLICATED 2 PORQUE SECUENCIA ES LA MISMA PERO NAME ES DIFERENTE 
#df_not_duplicated=pd.concat([df_not_duplicated, df_data_duplicated_2])
df_not_duplicated
```

```python
#se borran valores nulos y se eliminan los duplicados sobrantes de los que se guarda una copia y se ordenan los datos
df_not_duplicated.dropna(subset=['sequence', 'half_life_seconds'], inplace=True)
df_not_duplicated.drop_duplicates(subset=['sequence', 'modifications'], keep='first', inplace=True)
df_not_duplicated.sort_index(inplace=True)
```

```python
df_not_duplicated
```

```python
df_filtered = df_not_duplicated[df_not_duplicated['half_life_seconds'].str.contains(r'^\d+\.?\d*\s*(hours|min|minutes)$')]
df_object = df_not_duplicated[~df_not_duplicated['half_life_seconds'].str.contains(r'^\d+\.?\d*\s*(hours|min|minutes)$')]
df_filtered.reset_index(drop=True, inplace=True)
df_object.reset_index(drop=True, inplace=True)
```

```python
df_filtered
```

```python
for i in range(len(df_filtered)):
    if 'hours' in df_filtered.loc[i, 'half_life_seconds']:
        df_filtered.loc[i, 'half_life_seconds'] = float(df_filtered.loc[i, 'half_life_seconds'].split(' ')[0]) * 3600
    else:
        df_filtered.loc[i, 'half_life_seconds'] = float(df_filtered.loc[i, 'half_life_seconds'].split(' ')[0]) * 60
```

```python
df_object
```

```python
#Por si se quieren quitar las modificaciones
df_filter = df_filtered[df_filtered['is_mod'] == 'False']
df_filter = df_filtered.drop(columns=['is_mod', 'modifications'])
df_filter_ob = df_object[df_object['is_mod'] == 'False']
df_filter_ob = df_object.drop(columns=['is_mod', 'modifications'])
```

```python
df_filtered=df_filtered.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
df_object=df_object.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
```

```python
#df_filter.to_csv('../datasets/plifepred.csv', index=False)
#df_filter_ob.to_csv('../datasets/plifepred_object.csv', index=False)
df_filtered.to_csv('PepTherDia.csv', index=False)
df_object.to_csv('PepTherDia_object.csv', index=False)
```
