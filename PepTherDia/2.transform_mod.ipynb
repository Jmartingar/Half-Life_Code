```python
import pandas as pd
import numpy as np
```

```python
def mod_col(df):
    for i in range(len(df)):
        df.loc[i,'modifications'] = f"{df.loc[i, 'N-terminus']}, {df.loc[i, 'C-terminus']}"
        df.loc[i,'is_mod'] = 'True'
```

```python
df = pd.read_csv('PepTherDia_original.csv')
```

```python
df.drop(columns=['ID'], inplace=True)
```

```python
df2 = df
mod_col(df2)
```

```python
df2.drop(['N-terminus', 'C-terminus'], axis=1, inplace=True)
df2.rename(columns={'Sequence': 'sequence', 'Terminal Half-Life': 'half_life_seconds'}, inplace=True)
```

```python
df2['sequence'] = df2['sequence'].str.upper()
df2.dropna(subset=['sequence', 'half_life_seconds'], inplace=True)
df2.drop_duplicates(subset=['sequence'], inplace=True)
```

```python
df_filtered = df2[df2['half_life_seconds'].str.contains(r'^\d+\.?\d*\s*(hours|min|minutes)$')]
df_object = df2[~df2['half_life_seconds'].str.contains(r'^\d+\.?\d*\s*(hours|min|minutes)$')]
df_filtered.reset_index(drop=True, inplace=True)
df_object.reset_index(drop=True, inplace=True)
```

```python
df_filtered
```

```python
for i in range(len(df_filtered)):
    if 'hours' in df_filtered.loc[i, 'half_life_seconds']:
        df_filtered.loc[i, 'half_life_seconds'] = float(df_filtered.loc[i, 'half_life_seconds'].split(' ')[0]) * 3600
    else:
        df_filtered.loc[i, 'half_life_seconds'] = float(df_filtered.loc[i, 'half_life_seconds'].split(' ')[0]) * 60
```

```python
df_object
```

```python
#Por si se quieren quitar las modificaciones
df_filter = df_filtered[df_filtered['is_mod'] == 'False']
df_filter = df_filtered.drop(columns=['is_mod', 'modifications'])
df_filter_ob = df_object[df_object['is_mod'] == 'False']
df_filter_ob = df_object.drop(columns=['is_mod', 'modifications'])
```

```python
df_filtered=df_filtered.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
df_object=df_object.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
```

```python
#df_filter.to_csv('../datasets/plifepred.csv', index=False)
#df_filter_ob.to_csv('../datasets/plifepred_object.csv', index=False)
df_filtered.to_csv('PepTherDia.csv', index=False)
df_object.to_csv('PepTherDia_object.csv', index=False)
```
