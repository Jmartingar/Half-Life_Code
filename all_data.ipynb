```python
import pandas as pd
```

```python
def canon_or_notcanon(sequence):
     alphabet = set("ACDEFGHIKLMNPQRSTVWY")
     sequence = sequence.strip()
     is_canon = True
     for res in set(sequence):
         if res not in alphabet:
             is_canon = False
     return is_canon
```

```python
canon= []
notcanon= []
```

```python
antiviral=pd.read_csv("Antiviral/antiviral.csv")
hlp_10=pd.read_csv("HLP/HLP_10.csv")
hlp_16=pd.read_csv("HLP/HLP_16.csv")
peplife=pd.read_csv("PEPLIFE/PEP_LIFE.csv")
peptherdia=pd.read_csv("PepTherDia/PepTherDia.csv")
plifepred=pd.read_csv("Plifepred/plifepred.csv")
serum=pd.read_csv("Serum/serum.csv")
```

```python
df_concat=pd.concat([antiviral,hlp_10,hlp_16,peplife,peptherdia,plifepred,serum])
```

```python
df_concat['is_mod'].value_counts()
```

```python
#dimensiones de datos duplicados
df_concat[df_concat.duplicated(subset=['sequence', 'modifications'], keep=False)].shape
```

```python
#dimensiones de datos no duplicados
df_concat[~df_concat.duplicated(subset=['sequence', 'modifications'], keep=False)].shape
```

```python
#separacion de datos duplicados que tengan la misma secuencia y modificaciones (se examinan las secuencia, si modificaiones y secuencia es igual pero half_life_seconds es diferente se borran todos los duplicados)
df_data_duplicated=(df_concat[df_concat.duplicated(subset=['sequence', 'modifications'], keep=False)]).sort_values(by='sequence')
df_data_duplicated
```

```python
#separacion de datos duplicados que tengan la misma secuencia, modificaciones y half_life_seconds (de esto solo se guarda una copia) 
df_data_duplicated_2=(df_concat[df_concat.duplicated(subset=['sequence', 'modifications', 'half_life_seconds'], keep=False)]).sort_values(by='sequence')
df_data_duplicated_2
```

```python
#separacion de datos no duplicados
df_not_duplicated=(df_concat[~df_concat.duplicated(subset=['sequence', 'modifications'], keep=False)]).sort_values(by='sequence')
#se concatenan los datos no duplicados con los datos duplicados pero que se mantiene una copia
df_not_duplicated=pd.concat([df_not_duplicated, df_data_duplicated_2])
df_not_duplicated
```

```python
#se borran valores nulos y se eliminan los duplicados sobrantes de los que se guarda una copia y se ordenan los datos
df_not_duplicated.dropna(subset=['sequence', 'half_life_seconds'], inplace=True)
df_not_duplicated.drop_duplicates(subset=['sequence', 'modifications'], keep='first', inplace=True)
df_not_duplicated.sort_index(inplace=True)
```

```python
df_not_duplicated
```

```python
df_not_duplicated.to_csv('all_data.csv', index=False)
```

```python
all_data_mod=df_not_duplicated[df_not_duplicated['is_mod']==True]
all_data_nomod=df_not_duplicated[df_not_duplicated['is_mod']==False]
```

```python
all_data_mod.to_csv('all_data_mod.csv',index=False)
all_data_nomod.to_csv('all_data_nomod.csv',index=False)
```

```python
for sequence in all_data_nomod["sequence"].unique():
    if canon_or_notcanon(sequence):
        canon.append(sequence)
    else:
        notcanon.append(sequence)
```

```python
all_data_canon = all_data_nomod[all_data_nomod["sequence"].isin(canon)]
all_data_notcanon = all_data_nomod[all_data_nomod["sequence"].isin(notcanon)]
```

```python
all_data_canon.to_csv('all_data_nomod_canon.csv',index=False, columns=["sequence","half_life_seconds", "experimental_characteristics"])
all_data_notcanon.to_csv('all_data_nomod_notcanon.csv',index=False, columns=["sequence","half_life_seconds", "experimental_characteristics"])
```
