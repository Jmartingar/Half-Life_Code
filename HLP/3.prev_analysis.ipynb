```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_10 = pd.read_csv('HLP_10.csv')
df_16 = pd.read_csv('HLP_16.csv')
df=pd.read_csv('HLP.csv')
```

```python
df_10.shape, df_16.shape
```

```python
df_10.describe(), df_16.describe()
```

```python
df_concat=pd.concat([df_10,df_16],axis=0)
```

```python
sns.histplot(df_concat['half_life_seconds'], bins=10, color='skyblue', kde=False)

plt.xlabel('Vida media (segundos)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Vida Media (HLP_10 y HLP_16)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='half_life_seconds', data=df_10)
plt.title('Boxplot de Vida Media (HLP10)')
plt.xlabel('Vida media (segundos)')
plt.show()
```

```python
sns.histplot(df_16['half_life_seconds'], bins=10, color='skyblue', kde=False)

plt.xlabel('Vida media (segundos)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Vida Media (HLP_16)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='half_life_seconds', data=df_16)
plt.title('Boxplot de Vida Media (HLP16)')
plt.xlabel('Vida media (segundos)')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso_forest = IsolationForest(contamination=0.10, random_state=10)
df['is_outlier'] = iso_forest.fit_predict(df[['half_life_seconds']])
df_no_outliers = df[df['is_outlier'] == 1] 
print(df_no_outliers[['half_life_seconds']].describe())
```
