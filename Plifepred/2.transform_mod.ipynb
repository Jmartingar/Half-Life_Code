```python
import pandas as pd
```

```python
def categorize_hl(row):
    if row <= 7200:
        return 'Low'
    elif 7200 < row <= 18000:
        return 'Medium'
    else:
        return 'High'
```

```python
def mod_false(df):
    for i in range(len(df)):
        if df.loc[i,'nter'] != 'Free' or df.loc[i,'cter'] != 'Free':
            df.loc[i,'is_mod'] = 'True'
            df.loc[i,'modifications'] = f"{df.loc[i, 'nter']}, {df.loc[i, 'cter']}"
        else:
            df.loc[i,'is_mod'] = 'False'
            df.loc[i,'modifications'] = ''
```

```python
df = pd.read_csv('plife_mod.csv')
df2 = pd.read_csv('plife_freemod.csv')
```

```python
df2.rename(columns={'hl in seconds': 'Half-life in seconds'}, inplace=True)
```

```python
df3=pd.DataFrame()
df3=pd.concat([df,df2],ignore_index=True)
df3.drop_duplicates(keep='first',inplace=True)
df3.reset_index(drop=True, inplace=True)
mod_false(df3)
```

```python
df3['modifications'] = df3['modifications'].replace([', Free', 'Free,'], '', regex=True)
df3 = df3.rename(columns={'seq': 'sequence', 'Half-life in seconds': 'half_life_seconds'})
```

```python
df3
```

```python
df3['sequence'] = df3['sequence'].str.upper()
```

```python
#dimensiones de datos duplicados
df3[df3.duplicated(subset=['sequence', 'modifications'], keep=False)].shape
```

```python
#dimensiones de datos no duplicados
df3[~df3.duplicated(subset=['sequence', 'modifications'], keep=False)].shape
```

```python
#separacion de datos duplicados que tengan la misma secuencia y modificaciones (se examinan las secuencia, si modificaiones y secuencia es igual pero half_life_seconds es diferente se borran todos los duplicados)
df_data_duplicated=(df3[df3.duplicated(subset=['sequence', 'modifications'], keep=False)]).sort_values(by='sequence')
df_data_duplicated
```

```python
#separacion de datos duplicados que tengan la misma secuencia, modificaciones y half_life_seconds (de esto solo se guarda una copia) 
df_data_duplicated_2=(df3[df3.duplicated(subset=['sequence', 'modifications', 'half_life_seconds'], keep=False)]).sort_values(by='sequence')
df_data_duplicated_2
```

```python
#separacion de datos no duplicados
df_not_duplicated=(df3[~df3.duplicated(subset=['sequence', 'modifications'], keep=False)]).sort_values(by='sequence')
#se concatenan los datos no duplicados con los datos duplicados pero que se mantiene una copia
df_not_duplicated=pd.concat([df_not_duplicated, df_data_duplicated_2])
df_not_duplicated
```

```python
#se borran valores nulos y se eliminan los duplicados sobrantes de los que se guarda una copia y se ordenan los datos
df_not_duplicated.dropna(subset=['sequence', 'half_life_seconds'], inplace=True)
df_not_duplicated.drop_duplicates(subset=['sequence', 'modifications'], keep='first', inplace=True)
df_not_duplicated.sort_index(inplace=True)
```

```python
df_not_duplicated
```

```python
df_not_duplicated=df_not_duplicated.filter(['sequence', 'half_life_seconds', 'is_mod', 'modifications'])
```

```python
#Por si se quieren quitar las modificaciones
df_filtered = df3[df3['is_mod'] == 'False']
df_filtered = df_filtered.drop(columns=['is_mod', 'modifications'])
```

```python
df_not_duplicated['hl_category'] = df_not_duplicated['half_life_seconds'].apply(categorize_hl)
```
